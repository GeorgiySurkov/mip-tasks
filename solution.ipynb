```python
import numpy as np
import matplotlib.pyplot as plt
```

# Задача 1

Мы хотим найти полином пятого порядка $\( s(t) \)$, который удовлетворяет следующим условиям:

1. $\( s(0) = 0 \)$
2. $\( s(T) = 1 \)$
3. $\( \dot{s}(0) = 0 \)$
4. $\( \dot{s}(T) = 0 \)$
5. $\( \ddot{s}(0) = 0 \)$
6. $\( \ddot{s}(T) = 0 \)$

Полином пятого порядка:

$\[ s(t) = a_0 + a_1 t + a_2 t^2 + a_3 t^3 + a_4 t^4 + a_5 t^5 \]$

Граничные условия

1. $\( s(0) = a_0 = 0 \)$
2. $\( s(T) = a_0 + a_1 T + a_2 T^2 + a_3 T^3 + a_4 T^4 + a_5 T^5 = 1 \)$
3. $\( \dot{s}(0) = a_1 = 0 \)$
4. $\( \dot{s}(T) = a_1 + 2a_2 T + 3a_3 T^2 + 4a_4 T^3 + 5a_5 T^4 = 0 \)$
5. $\( \ddot{s}(0) = 2a_2 = 0 \)$
6. $\( \ddot{s}(T) = 2a_2 + 6a_3 T + 12a_4 T^2 + 20a_5 T^3 = 0 \)$

Остается решить 3 уравнения

1. $\( a_3 T^3 + a_4 T^4 + a_5 T^5 = 1 \)$
2. $\( 3a_3 T^2 + 4a_4 T^3 + 5a_5 T^4 = 0 \)$
3. $\( 6a_3 T + 12a_4 T^2 + 20a_5 T^3 = 0 \)$


```python
T = 1

A = np.array([
    [T**3, T**4, T**5],
    [3*T**2, 4*T**3, 5*T**4],
    [6*T, 12*T**2, 20*T**3]
])
b = np.array([1, 0, 0])

a3, a4, a5 = np.linalg.solve(A, b)

def s(t):
    return a3*t**3 + a4*t**4 + a5*t**5

def ds(t):
    return 3*a3*t**2 + 4*a4*t**3 + 5*a5*t**4

def dds(t):
    return 6*a3*t + 12*a4*t**2 + 20*a5*t**3

t = np.linspace(0, T, 1000)

s_t = s(t)
ds_t = ds(t)
dds_t = dds(t)

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, s_t, label='s(t)')
plt.title('s(t)')
plt.xlabel('Time (s)')
plt.ylabel('s(t)')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, ds_t, label='ds(t)', color='orange')
plt.title('ds(t)')
plt.xlabel('Time (s)')
plt.ylabel('ds(t)')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, dds_t, label='dds(t)', color='green')
plt.title('dds(t)')
plt.xlabel('Time (s)')
plt.ylabel('dds(t)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

# Задача 2

Интерполяция траектории с использованием полинома седьмого порядка, который удовлетворяет условиям:

1. $\( s(0) = 0 \)$
2. $\( s(T) = 1 \)$
3. $\( \dot{s}(0) = 0 \)$
4. $\( \dot{s}(T) = 0 \)$
5. $\( \ddot{s}(0) = 0 \)$
6. $\( \ddot{s}(T) = 0 \)$
7. $\( \dddot{s}(0) = 0 \)$
8. $\( \dddot{s}(T) = 0 \)$

Полином седьмого порядка:

$\[ s(t) = a_0 + a_1 t + a_2 t^2 + a_3 t^3 + a_4 t^4 + a_5 t^5 + a_6 t^6 + a_7 t^7 \]$

Граничные условия

1. $\( s(0) = a_0 = 0 \)$
2. $\( s(T) = a_0 + a_1 T + a_2 T^2 + a_3 T^3 + a_4 T^4 + a_5 T^5 + a_6 T^6 + a_7 T^7 = 1 \)$
3. $\( \dot{s}(0) = a_1 = 0 \)$
4. $\( \dot{s}(T) = a_1 + 2a_2 T + 3a_3 T^2 + 4a_4 T^3 + 5a_5 T^4 + 6a_6 T^5 + 7a_7 T^6 = 0 \)$
5. $\( \ddot{s}(0) = 2a_2 = 0 \)$
6. $\( \ddot{s}(T) = 2a_2 + 6a_3 T + 12a_4 T^2 + 20a_5 T^3 + 30a_6 T^4 + 42a_7 T^5 = 0 \)$
7. $\( \dddot{s}(0) = 6a_3 = 0 \)$
8. $\( \dddot{s}(T) = 6a_3 + 24a_4 T + 60a_5 T^2 + 120a_6 T^3 + 210a_7 T^4 = 0 \)$

Остается решить систему из 4 уравнений

1. $\( a_4 T^4 + a_5 T^5 + a_6 T^6 + a_7 T^7 = 1 \)$
2. $\( 4a_4 T^3 + 5a_5 T^4 + 6a_6 T^5 + 7a_7 T^6 = 0 \)$
3. $\( 12a_4 T^2 + 20a_5 T^3 + 30a_6 T^4 + 42a_7 T^5 = 0 \)$
4. $\( 24a_4 T + 60a_5 T^2 + 120a_6 T^3 + 210a_7 T^4 = 0 \)$


```python
T = 1

A = np.array([
    [T**4, T**5, T**6, T**7],
    [4*T**3, 5*T**4, 6*T**5, 7*T**6],
    [12*T**2, 20*T**3, 30*T**4, 42*T**5],
    [24*T, 60*T**2, 120*T**3, 210*T**4]
])
b = np.array([1, 0, 0, 0])

a4, a5, a6, a7 = np.linalg.solve(A, b)

def s(t):
    return a4*t**4 + a5*t**5 + a6*t**6 + a7*t**7

def ds(t):
    return 4*a4*t**3 + 5*a5*t**4 + 6*a6*t**5 + 7*a7*t**6

def dds(t):
    return 12*a4*t**2 + 20*a5*t**3 + 30*a6*t**4 + 42*a7*t**5

def ddds(t):
    return 24*a4*t + 60*a5*t**2 + 120*a6*t**3 + 210*a7*t**4

t = np.linspace(0, T, 1000)

s_t = s(t)
ds_t = ds(t)
dds_t = dds(t)
ddds_t = ddds(t)

plt.figure(figsize=(12, 10))

plt.subplot(4, 1, 1)
plt.plot(t, s_t, label='s(t)')
plt.title('s(t)')
plt.xlabel('Time (s)')
plt.ylabel('s(t)')
plt.grid(True)
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(t, ds_t, label='ds(t)', color='orange')
plt.title('ds(t)')
plt.xlabel('Time (s)')
plt.ylabel('ds(t)')
plt.grid(True)
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(t, dds_t, label='dds(t)', color='green')
plt.title('dds(t)')
plt.xlabel('Time (s)')
plt.ylabel('dds(t)')
plt.grid(True)
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(t, ddds_t, label='ddds(t)', color='red')
plt.title('ddds(t)')
plt.xlabel('Time (s)')
plt.ylabel('ddds(t)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
